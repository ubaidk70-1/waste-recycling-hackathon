```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import os 
```

```python
# Define the path to the raw data file

file_path = os.path.join('..', 'data', 'raw', 'Waste_Management_and_Recycling_India.csv')

```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first 5 rows to confirm it's loaded
print("Dataset loaded successfully!")
df.head()
```

```python
# Get a concise summary of the dataframe
df.info()
```

```python
# Generate descriptive statistics for numerical columns
df.describe()
```

```python
# Generate descriptive statistics for numerical columns and counts for categorical columns
df.describe(include='all')
```

```python
# See how many unique values are in each column
df.nunique()
```

```python
# Univariate Analysis (Analyzing single variables)

# Visualizing the Target Variable

sns.set_style('whitegrid')

# Create a histogram for the 'Recycling Rate (%)'
plt.figure(figsize=(10, 6)) 
sns.histplot(df['Recycling Rate (%)'], bins=20, kde=True)

plt.title('Distribution of Recycling Rate (%)', fontsize=16)
plt.xlabel('Recycling Rate (%)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Recycling rates are fairly evenly distributed around the center.
# There are no obvious outliers.

# Now that we understand our target variable on its own, let's see how it's affected by a categorical feature
```

```python
# Create a box plot to compare Recycling Rate by Waste Type
plt.figure(figsize=(12, 7)) 
sns.boxplot(x='Waste Type', y='Recycling Rate (%)', data=df)

plt.title('Recycling Rate (%) by Waste Type', fontsize=16)
plt.xlabel('Waste Type', fontsize=12)
plt.ylabel('Recycling Rate (%)', fontsize=12)
plt.show()
```

**The key insight here is that Waste Type, by itself, doesn't seem to have a strong influence on the recycling rate.** Whether the waste is Plastic, Organic, or E-Waste, the recycling rates tend to fall within the same general range. This is a crucial finding for our project!

```python
# Numerical Features:

# Create histograms for Waste Generated (Tons/Day), Population Density, Cost of Waste Management, and Landfill Capacity.

# Set up the figure for 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Distribution of Key Numerical Features', fontsize=20)

# Plot 1: Waste Generated (Tons/Day)
sns.histplot(df['Waste Generated (Tons/Day)'], ax=axes[0, 0], kde=True)
axes[0, 0].set_title('Waste Generated (Tons/Day)')

# Plot 2: Population Density (People/km²)
sns.histplot(df['Population Density (People/km²)'], ax=axes[0, 1], kde=True)
axes[0, 1].set_title('Population Density (People/km²)')

# Plot 3: Cost of Waste Management (₹/Ton)
sns.histplot(df['Cost of Waste Management (₹/Ton)'], ax=axes[1, 0], kde=True)
axes[1, 0].set_title('Cost of Waste Management (₹/Ton)')

# Plot 4: Landfill Capacity (Tons)
sns.histplot(df['Landfill Capacity (Tons)'], ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Landfill Capacity (Tons)')

# Adjust layout to prevent titles from overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Categorical Features:
# Create count plots (bar charts) for Waste Type, Municipal Efficiency Score, and Disposal Method.

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Distribution of Key Categorical Features', fontsize=20)

# Plot 1: Waste Type
sns.countplot(ax=axes[0], x='Waste Type', data=df)
axes[0].set_title('Waste Type Counts')
axes[0].tick_params(axis='x', rotation=45) # Rotate labels for better readability

# Plot 2: Disposal Method
sns.countplot(ax=axes[1], x='Disposal Method', data=df)
axes[1].set_title('Disposal Method Counts')
axes[1].tick_params(axis='x', rotation=45)

# Adjust layout and show the plot
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**Insight of Univariate Analysis**

Our analysis of individual features revealed key characteristics of the dataset. 
+ The numerical features, such as Population Density and Landfill Capacity, show multimodal distributions, suggesting the data contains natural clusters, like different types of cities or landfill sizes. Other numerical features like Waste Generated are more uniformly distributed. 
+ For the categorical features, we found that both Waste Type and Disposal Method are perfectly balanced. This is excellent for modeling as it ensures our algorithm can learn from each category equally without bias.

```python
# Bivariate Analysis (Analyzing relationships

# List of numerical features to plot against the target
numerical_features = [
    'Waste Generated (Tons/Day)',
    'Population Density (People/km²)',
    'Cost of Waste Management (₹/Ton)',
    'Awareness Campaigns Count',
    'Landfill Capacity (Tons)'
]

# Set up the figure for subplots
# We need 5 plots, so a 3x2 grid is suitable
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
fig.suptitle('Numerical Features vs. Recycling Rate (%)', fontsize=20)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each numerical feature and create a scatter plot
for i, feature in enumerate(numerical_features):
    sns.scatterplot(x=feature, y='Recycling Rate (%)', data=df, ax=axes[i], alpha=0.5)
    axes[i].set_title(f'Recycling Rate vs. {feature}', fontsize=12)

# Hide the empty subplot (since we have 5 plots in a 3x2 grid)
axes[-1].set_visible(False)

# Adjust layout to prevent titles from overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
# List of categorical features to plot
categorical_features = [
    'Waste Type',
    'Municipal Efficiency Score (1-10)',
    'Disposal Method'
]

# Set up the figure for a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Categorical Features vs. Recycling Rate (%)', fontsize=20)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each categorical feature and create a box plot
for i, feature in enumerate(categorical_features):
    sns.boxplot(x=feature, y='Recycling Rate (%)', data=df, ax=axes[i])
    axes[i].set_title(f'Recycling Rate vs. {feature}', fontsize=14)
    axes[i].tick_params(axis='x', rotation=15) # Rotate labels slightly if needed

# Hide the last (empty) subplot
axes[-1].set_visible(False)

# Adjust layout to prevent titles from overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```

```python
# Correlation Heatmap of every Numerical columns.

# Create a correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr(numeric_only=True)

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

**Insight of Bivariate Analysis**

Our bivariate analysis revealed that there are no strong, simple linear relationships between the numerical features and the Recycling Rate (%), as confirmed by both the correlation heatmap and scatter plots. The relationships are more complex. Box plots showed that categorical features like Waste Type and Disposal Method have little impact on their own. However, the Municipal Efficiency Score showed a significant, non-linear relationship with the recycling rate, making it a key predictive feature for our model.

#### Final Summary of Exploratory Data Analysis (EDA)

This Exploratory Data Analysis was conducted to understand the fundamental characteristics of the waste management dataset and to uncover initial patterns that will guide our modeling strategy.

**1. Data Quality and Structure:**

  + The dataset is of high quality, containing no missing values, which allows us to proceed directly to analysis without needing to clean or impute data.

  + Categorical features like Waste Type and Disposal Method were found to be perfectly balanced, which is highly advantageous as it prevents model bias towards any single category.

**2. Key Insights from Univariate Analysis:**

  + Our target variable, Recycling Rate (%), is normally distributed, which is ideal for regression tasks.

  + Numerical features like Population Density and Landfill Capacity are multimodal, suggesting that our dataset contains distinct clusters of cities (e.g., megacities vs. smaller cities) or landfill types.

**3. Key Insights from Bivariate Analysis:**

  + Correlation analysis revealed **no strong linear relationships** between any single numerical feature and the recycling rate. This indicates that simple linear models will likely be insufficient.

  + The most significant finding is the **strong, non-linear relationship** between the Municipal Efficiency Score and the Recycling Rate (%). While not a simple straight-line trend, different scores correspond to clearly different distributions of recycling rates, making it a crucial predictive feature.

  + Other features, like Waste Type and Cost of Management, showed no discernible relationship with the recycling rate on their own.

**Conclusion:**

The EDA concludes that predicting the recycling rate is a complex, non-linear problem. The solution will not come from a single feature but from the interactions between multiple features. Therefore, our modeling strategy should focus on powerful, tree-based algorithms like XGBoost or LightGBM, which are specifically designed to capture these intricate patterns. The next step is to preprocess the data and engineer new features to prepare it for these advanced models.
